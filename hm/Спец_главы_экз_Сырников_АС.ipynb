# Задание 13

Сколькими способами можно распределить 10 одинаковых яблок между 4 детьми, если каждый ребенок должен получить хотя бы одно яблоко? 

# Решение

Пусть яблоки разложены в линию и между ними расположено 3 перегородки. Все яблоки слева от первой отдаются первому ребенку, справа от первой и слева от второй - второму и т.д. Перегородки не могут стоять по краям и между ними должно быть хотя бы по одному яблоку (так каждый получит хотя-бы одно яблоко).

Задача сводится к расстановке 3 перегородок на 9 мест между яблоками.

$
C_{9}^{3} = \dfrac{9!}{3! \cdot (9-3)!} = \dfrac{9!}{3! \cdot 6!} = \dfrac{7 \cdot 8 \cdot 9}{1 \cdot 2 \cdot 3} = 84
$

# Задание 15

Монету бросают 6 раз. Какова вероятность того, что орел выпадет ровно 4 раза? 

# Решение

По формуле биноминального распределения:

$
P(k, n, p) = C_n^k \cdot p^k \cdot (1 - p)^{n-k} = C_6^4 \cdot 0.5^4 \cdot 0.5^{2} = \dfrac{6!}{4!(6-4)!} \cdot \dfrac{1}{64} = \dfrac{15}{64} \approx 0.23
$

Ответ: $\dfrac{15}{64} \approx 0.23$

# Задание 16

Дана функция:

$
f(x)=\sqrt{\dfrac{\ln \left(x^{2}-1\right)}{x-2}}
$

Найдите область определения функции.

Найдите область значений функции.

Определите, является ли функция четной или нечетной. 

# Решение

Найдем область определения:

$
\left\{\begin{array}{l}
x^{2}-1 > 0& \text{при}& x \in (-\infty; -1) \cup (1; +\infty) \\
x - 2 \neq 0& \text{при}& x \in (-\infty; 2) \cup (2; +\infty) \\
\dfrac{\ln \left(x^{2}-1\right)}{x-2} \geq 0
\end{array}\right.
$

$\dfrac{\ln \left(x^{2}-1\right)}{x-2} \geq 0$

По методу рационализации:

$\dfrac{x^{2}-2}{x-2} \geq 0$

$x \in [-\sqrt{2}; \sqrt{2}] \cup (2; +\infty)$

Тогда область определения:

$
D_f = [-\sqrt{2}; -1) \cup (1;\sqrt{2}] \cup (2; +\infty)
$

Область значений $E_f = [0; +\infty)$ из-за наличчия корня.

Функция четная если $f(x) = f(-x)$

$f(-x) = \sqrt \frac {\ln(x^2-1)}{-x-2} \neq f(x)$

Функция нечетная если $f(x) = -f(-x)$

$-f(-x) = -\sqrt \frac {\ln(x^2-1)}{-x-2} \neq f(x)$

Следовательно функция не четная и не нечетная.

# Задание 17

В треугольнике ABC известно, что AC = BC, AH - высота, tg(BAC) = 7/24. Найдите косинус угла BAH. 



# Решение

$AC=BC \implies$ ABC равнобедренный $\implies\tan{BAC} = \tan{ABH} = \dfrac{AH}{BH} \implies AH = \tan{BAC} \cdot BH$

По теореме пифагора:

$AB = \sqrt{AH^2 + BH^2}$

$\cos{BAH} = \dfrac{AH}{AB} = \dfrac{AH}{\sqrt{AH^2 + BH^2}} = \dfrac{\tan{BAC} \cdot BH}{\sqrt{(\tan{BAC} \cdot BH)^2 + BH^2}} = \dfrac{\tan{BAC} \cdot BH}{\sqrt{(\tan^2{BAC} + 1)\cdot BH^2}} = \dfrac{\tan{BAC}}{\sqrt{\tan^2{BAC} + 1}} = \dfrac{7}{24 \sqrt{(\dfrac{7}{24})^2 + 1}} = \dfrac{7}{24 \sqrt{\dfrac{49}{576} + 1}} =$

$=\dfrac{7}{24 \sqrt{\dfrac{625}{576}}} = \dfrac{7 \cdot 24}{24 \cdot 25} = \dfrac{7}{25}$

Ответ: $\dfrac{7}{25}$

# Задание 18

В треугольнике со сторонами 9 и 5 проведены высоты к этим сторонам. Высота, проведённая к первой из этих сторон, равна 10. Чему равна высота, проведённая ко второй стороне? 

# Решение



$S_{ABC} = \dfrac{1}{2}AC\cdot BD = \dfrac{1}{2}\cdot 9 \cdot 10 = 45$

$S_{ABC} = \dfrac{1}{2}AB\cdot CE =\dfrac{1}{2}\cdot 5 \cdot CE = \dfrac{5}{2} \cdot CE$

$CE = S_{ABC} \cdot \dfrac{2}{5} = 45 \cdot \dfrac{2}{5} = 18$

Ответ: 18

# Задание 20

В равнобедренной трапеции основания равны 12 и 27, острый угол равен 60 градусам.

Найдите её периметр. 

# Решение



$AB=CD, AD=27, BC=12$

$BE, CF$ - высоты

$\angle BAE = \angle CDF = 60 \implies \angle ABE = \angle DCF = 90 - 60 = 30$

Трапеция равнобедренная, тогда $AE=FD=\dfrac{AD - EF}{2} = \dfrac{AD - BC}{2} = \dfrac{27 - 12}{2}=7.5$

$AE$ лежит против угла 30 градусов, следовательно $AB=CD=2AE = 15$.

Периметр $P=AB+ CD + BC + EF + AE + FD =15 + 15 + 12 + 12 + 7.5 + 7.5 = 69$

Ответ: 69
