Задание 2 (про порошок)

Используем z-тест так как размер выборки больше 30 и известно стандартное отклонение.

Нулевая гипотеза: истинное среднее значение веса равно 3.00 кг.

Альтернативная гипотеза: истинное среднее значение веса не равно 3.00 кг.

```python
'''Ваш ответ на задание №2 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

sample_size = 50 # размер выборки
mean = 3.00 # предполагаемое среднее
sample_mean = 3.005 # выборочное среднее
std = 0.015 # стандартное отклонение
alpha = 0.05 # уровень значимости

# расчет z-статистики
z = (sample_mean - mean) / (std / np.sqrt(sample_size))

# вычисление критического значения
critical = stats.norm.ppf(1 - alpha / 2)

# вычисление p_value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z)))

# статистический вывод на основе сравнения критического значения с расчетным
if np.abs(z) > critical:
    z_result = 'Отвергаем нулевую гипотезу'
else:
    z_result = 'Не отвергаем нулевую гипотезу'

# статистический вывод на основе сравнения p-value и уровня значимости
if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# содержательный вывод
if 'Отвергаем нулевую гипотезу' in [z_result, p_value_result]:
    result = 'Есть достаточные доказательства, что истинное среднее значение не равно 3.00 кг. Машина требует настройки.'
else:
    result = 'Нет достаточных доказательств, что истинное среднее значение не равно 3.00 кг. Машина не требует настройки.'

# вывод результатов
print('z-статистика:', z)
print('Критическое значение:', critical)
print('p_value:', p_value)
print('Статистический вывод (z-тест):', z_result)
print('Статистический вывод (p_value)', p_value_result)
print('Содержательный вывод:', result)
```

Задание 3 (про двигатели)

Размер выборки меньше 30 и сравниваются средние значения двух независимых групп, используем t-тест (независимый t-критерий).

Нулевая гипотеза: средние прочностные характеристики двух веществ равны

Альтернативная гипотеза: средние прочностные характеристики двух веществ не равны

```python
'''Ваш ответ на задание №3 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

sample_mean_st = 90 # средняя прочность стальных компонентов
sample_mean_al = 88 # средняя прочность алюминиевых компонентов
std_st = 2.3 # стандартное отклонение стальных компонентов
std_al = 2.2 # стандартное отклонение алюминиевых компонентов
sample_size_st = 15 # размер выборки стальных компонентов
sample_size_al = 10 # размер выборки алюминиевых компонентов
alpha = 0.05 # уровень значимости

# расчет t-статистики
t = (sample_mean_st - sample_mean_al) / np.sqrt((std_st**2 / sample_size_st) + (std_al**2 / sample_size_al))

# степени свободы
df = sample_size_st + sample_size_al - 2

# вычисление критического значения
critical = stats.t.ppf(1 - alpha / 2, df)

# вычисление p_value
p_value = 2 * (1 - stats.t.cdf(np.abs(t), df))

# статистический вывод на основе сравнения критического значения с расчетным
if np.abs(t) > critical:
    t_result = "Отвергаем нулевую гипотезу"
else:
    t_result = "Не отвергаем нулевую гипотезу"

# статистический вывод на основе сравнения p-value и уровня значимости
if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# Содержательный вывод
if "Отвергаем нулевую гипотезу" in [t_result, p_value_result]:
    result = "Есть достаточные доказательства, что средние прочностные характеристики стальных и алюминиевых компонентов не равны. Нельзя рассматривать использование алюминиевых компонентов."
else:
    result = "Нет достаточных доказательств, что средние прочностные характеристики стальных и алюминиевых компонентов не равны. Следовательно, можно рассмотреть использование алюминиевых компонентов."

# вывод результатов
print("t-статистика:", t)
print("Критическое значение:", critical)
print("p_value:", p_value)
print("Статистический вывод (t-тест):", t_result)
print("Статистический вывод (p_value):", p_value_result)
print("Содержательный вывод:", result)
```

Задание 4 (про токи)

Размер выборки меньше 30 и выборки независимы, используем t-тест (независимый t-критерий) учитывая условие задания о равной дисперсии.

Нулевая гипотеза: средний ток, пропускаемый новой схемой, равен среднему току старой схемы

Альтернативная гипотеза: средние токи отличаются

```python
'''Ваш ответ на задание №4 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

alpha = 0.05 # уровень значимости
I_old = [80.1, 82.3, 84.1, 82.6, 85.3, 81.3, 83.2, 81.7, 82.2, 81.4]  # ток старой к.
I_new = [80.7, 81.3, 84.6, 81.7, 86.3, 84.3, 83.7, 84.7, 82.8, 84.4, 85.2, 84.9] # ток новой к.

# расчет статистик
res = stats.ttest_ind(I_old, I_new)
t, p_value = res[0], res[1]

# вычисление критического значения
critical = stats.t.ppf(1 - alpha / 2, df = res.df)

# статистический вывод на основе сравнения критического значения с расчетным
if np.abs(t) > critical:
    t_result = "Отвергаем нулевую гипотезу"
else:
    t_result = "Не отвергаем нулевую гипотезу"

# статистический вывод на основе сравнения p-value и уровня значимости
if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# содержательный вывод
if "Отвергаем нулевую гипотезу" in [t_result, p_value_result]:
    result = "Есть достаточные доказательства, что средний ток, пропускаемый старой и новой схемами, различается. Не следует заменять старую конструкцию схемы на новую."
else:
    result = "Нет достаточных доказательств, что средний ток, пропускаемый старой и новой схемами, различается. Можно заменить старую конструкцию схемы на новую."

# вывод результатов
print("t-статистика:", t)
print("Критическое значение:", critical)
print("p_value:", p_value)
print("Статистический вывод (t-тест):", t_result)
print("Статистический вывод (p_value):", p_value_result)
print("Содержательный вывод:", result)
```

Нулевая гипотеза: средний уровень глюкозы в крови пациентов до и после приема препарата одинаков

Альтернативная гипотеза: средний уровень глюкозы в крови до и после приема препарата отличается

Нулевая гипотеза: средние напряжения разрушения нового и старого кабелей одинаковы

Альтернативная гипотеза: средние напряжения разрушения нового и старого кабелей различаются

```python
'''Ваш ответ на задание №6 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

alpha = 0.01 # уровень значимости
old_cable = [90.1, 92.2, 94.3, 92.1, 93.4, 91.1, 93.2, 91.5] # напряжения разрушения старого кабеля
new_cable = [92.7, 91.6, 94.7, 93.7, 96.5, 94.3, 93.7, 96.8, 98.9, 99.9]  # напряжения разрушения нового кабеля

# расчет статистик
res = stats.ttest_ind(old_cable, new_cable, equal_var=False)
t, p_value = res[0], res[1]

# вычисление критического значения
critical = stats.t.ppf(1 - alpha, df = res.df)

# статистический вывод на основе сравнения критического значения с расчетным
if np.abs(t) > critical:
    t_result = "Отвергаем нулевую гипотезу"
else:
    t_result = "Не отвергаем нулевую гипотезу"

# статистический вывод на основе сравнения p-value и уровня значимости
if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

if "Отвергаем нулевую гипотезу" in [t_result, p_value_result]:
    result = "Есть достаточные доказательства, что средние напряжения разрушения нового и старого кабелей различаются. Изготовитель должен заменить старый кабель на новый."
else:
    result = "Нет достаточных доказательств, что средние напряжения разрушения нового и старого кабелей различаются. Изготовитель не должен заменять старый кабель на новый."

# вывод результатов
print("t-статистика:", t)
print("Критическое значение:", critical)
print("p_value:", p_value)
print("Статистический вывод (t-тест):", t_result)
print("Статистический вывод (p_value):", p_value_result)
print("Содержательный вывод:", result)
```

Задание 8 (про фрукты и IQ)

Требуется сравнить дисперсии двух выборок, используем F-тест.

Нулевая гипотеза: дисперсии обеих групп одинаковы

Альтернативная гипотеза: дисперсии групп различаются


```python
'''Ваш ответ на задание №8 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

alpha = 0.05 # уровень значимости
oranges = [125, 138, 112, 130, 124, 125, 110, 112, 117, 126] # результаты питавшихся только апельсинами
pears = [124, 102, 142, 130, 131, 124, 127, 115, 108, 109] # результаты питавшихся только грушами

# вычисление дисперсий
oranges_var = np.var(oranges, ddof=1)
pears_var = np.var(pears, ddof=1)

# расчет статистик
#f = pears_var / oranges_var
f = oranges_var / pears_var
df_oranges, df_pears = len(oranges) - 1, len(pears) - 1  # степени свободы

# вычисление критического значения
critical = stats.f.ppf(1 - alpha, df_oranges, df_pears)

# вычисление p_value
p_value = 2 * min(stats.f.cdf(f, df_oranges, df_pears), 1 - stats.f.cdf(f, df_oranges, df_pears))

# cтатистический вывод на основе сравнения критического значения с расчетным
if f > critical or f < 1 / critical:
    f_result = "Отвергаем нулевую гипотезу"
else:
    f_result = "Не отвергаем нулевую гипотезу"

# статистический вывод на основе сравнения p-value и уровня значимости
if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# cодержательный вывод
if "Отвергаем нулевую гипотезу" in [f_result, p_value_result]:
    result = "Есть достаточные доказательства, что дисперсии обеих групп различаются."
else:
    result = "Нет достаточных доказательств, что дисперсии обеих групп различаются."

# вывод результатов
print("F-статистика:", t)
print("Критическое значение:", critical)
print("p_value:", p_value)
print("Статистический вывод (F-тест):", t_result)
print("Статистический вывод (p_value):", p_value_result)
print("Содержательный вывод:", result)
```

Нулевая гипотеза: распределение данных соответствует геометрическому распределению.

Альтернативная гипотеза: распределение данных не соответствует геометрическому распределению.

```python
'''Ваш ответ на задание №9 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

alpha = 0.05 # уровень значимости
x = np.arange(1, 10) # количество вылетов
observed_frequencies = np.array([36, 23, 11, 13, 8, 3, 2, 1, 3]) # количество самолетов по вылетам
n = np.sum(observed_frequencies)  # общее количество самолетов

# оценка параметра p
mean_value = np.sum(observed_frequencies * x) / n
p = 1 / mean_value

# вычисление ожидаемых частот
expected_frequencies = n * stats.geom.pmf(x, p)
expected_frequencies *= n / np.sum(expected_frequencies)

# расчет статистик
chi2, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

# критическое значение
df = len(observed_frequencies) - 1  # степени свободы
critical = stats.chi2.ppf(1 - alpha, df)

# статистический вывод
if chi2 > critical:
    chi2_result = "Отвергаем нулевую гипотезу"
else:
    chi2_result = "Не отвергаем нулевую гипотезу"

if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# cодержательный вывод
if "Отвергаем нулевую гипотезу" in [chi2_result, p_value_result]:
    result = "Есть достаточные доказательства того, что данные не соответствуют геометрическому распределению. Предположение о геометрическом распределении не подтверждается."
else:
    result = "Нет достаточных доказательств того, что данные не соответствуют геометрическому распределению. Предположение о геометрическом распределении подтверждается."

# Вывод результатов
print("Оценка λ:", mean_value)
print("Хи-квадрат-статистика:", chi2)
print("Критическое значение:", critical)
print("p-значение:", p_value)
print("Статистический вывод (Хи-квадрат-тест):", chi2_result)
print("Статистический вывод (p-значение):", p_value_result)
print("Содержательный вывод:", result)
```

Задание 11 (про тигров)

Мы имеем несколько групп (4 типа корма) и хотим проверить, есть ли статистически значимые различия между средними массами тигров для разных типов корма, будет применяться однофакторный тест ANOVA.

Нулевая гипотеза: средние массы тигров для всех типов корма одинаковы.

Альтернативная гипотеза: средние массы тигров для хотя бы одного типа корма отличаются.

```python
'''Ваш ответ на задание №11
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- в случае, если нулевая гипотеза отвергнута, сделан тест Тьюки для подробного сравнения средних
- сделано дополнение к статистическому выводу по результатам теста Тьюки
- сделан содержательный вывод по поставленной проблеме'''

alpha = 0.05 # уровень значимости
group1 = np.array([151, 135, 137, 118, 132, 135, 131, 121, 137, 140, 152, 133, 151, 132]) # масса тигров с кормом №1
group2 = np.array([108, 123, 145, 103, 110, 124, 116, 108, 104, 130, 106, 105, 114, 108]) # масса тигров с кормом №2
group3 = np.array([140, 152, 149, 132, 124, 146, 156, 137, 124, 145, 147, 139, 157, 139]) # масса тигров с кормом №3
group4 = np.array([127, 135, 110, 116, 119, 123, 102, 121, 118, 120, 114, 134, 121, 109]) # масса тигров с кормом №4
data = [group1, group2, group3, group4] # объединённые данные

# Выполнение ANOVA
f, p_value = stats.f_oneway(*data)

# Результаты ANOVA

print("F-статистика:", f)
print("p_value:", p_value)

# Статистический вывод по результатам ANOVA
if p_value < alpha:
    print("Отвергаем нулевую гипотезу.")
    # Если гипотеза отвергнута, выполняем тест Тьюки
    t_res = stats.tukey_hsd(*data)
    print(t_res)
else:
    print("Не отвергаем нулевую гипотезу.")
```

Тест Тьюки показал, что значимые различия между средним весом между группами, принимавшими корма 1 и 3, а также 2 и 4, отсутствуют. При сравнении групп тигров в других сочетаниях существуют значимые отличия.
